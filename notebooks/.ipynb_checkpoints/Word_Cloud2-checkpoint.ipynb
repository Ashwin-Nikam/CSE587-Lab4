```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd
```

```python
f = open("../outputs/output1/part-r-00000")
print(f.read())
```

Reading the output which has been generated using MapReduce

```python
table = pd.read_table("../outputs/output1/part-r-00000", sep = "\t")
```

Here we need to specify the path of the output file from MapReduce for WordCount which is in outputs/output1/ folder

```python
table.columns = ['word','count']
print table
```

```python
array = {}
for a, x in table.values:
    array[a] = x

wordcloud = WordCloud(background_color='white', width=2000,
                          height=2000)
wordcloud.generate_from_frequencies(frequencies = array)
plt.figure(figsize=(20,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

WordCloud generated using hashtags and mentions related to soccer
